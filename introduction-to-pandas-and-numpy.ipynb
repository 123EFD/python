loc() ;  = label-based , specify rows and columns based on their **index** 
iloc() : integer position-based, sp we have to specify rows and columns by their **integer** position values

**Value**
for example: df.loc['zero'] for loc (label)
             df.iloc[0] for iloc (position)

**List**
for example: df.loc[['zero','two]] for loc 
             df.iloc[[0,2]] for iloc 

**Slicing**
for example: df.loc['zero':'two] for loc > 0-2
             df.iloc[0:2] for iloc > 0-1 ( 2 is excluded )

**Lambda function**

function: use only once, to simplify the function block into linear 

```python
#basic functions
def sum_values(a,b):
    x = a+b
    return x
```

```python
sum_values(2,3)
```

```python
#lambda function (one linear)
#a,b=input;  a+b = output
#
sum_values_lambda = lambda a,b:a+b
sum_values_lambda(2,3)
```
